```python
data = ["Apple", "Mango", "Strawberry"]
with open(r"F:\Data_Analytics\Python\File_handling\fruits.txt", "w") as f:
    for i in data:
        f.write("\n"+i)
```

```python
with open(r"F:\Data_Analytics\Python\File_handling\fruits.txt", "r") as f:
    print(f.read())
```

```python
data = ["\nApple", "\nMango", "\nStrawberry"]
with open(r"F:\Data_Analytics\Python\File_handling\fruits.txt", "w") as f:
    for i in data:
        f.writelines("\n"+i)
```

json file

```python
import json
data = [
    {
    "name": "Vachas",
    "age":30,
    "city":"Delhi"
    }
]
with open(r"F:\Data_Analytics\Python\File_handling\employees.json", "w") as f:
    json.dump(data, f, indent=4)
```

```python
with open(r"F:\Data_Analytics\Python\File_handling\employees.json", "r") as f:
    print(json.load(f))
```

```python
import json
data = [
    {
    "name": "Vachas",
    "age":30,
    "city":"Delhi"
    }
]
data1 = {
    "name": "Priya",
    "age":31,
    "city":"Delhi"
    }
data.append(data1)
with open(r"F:\Data_Analytics\Python\File_handling\employees.json", "w") as f:
    json.dump(data, f, indent=4)
```

```python
with open(r"F:\Data_Analytics\Python\File_handling\employees.json", "r") as f:
    print(json.load(f))
```
