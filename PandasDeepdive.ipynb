Create a new column margin_type 
runs>=100 Huge  
runs>=50 Big  
runs>0 close  
else chasing

```python
def margin_type(runs):
    if runs>=100:
        return "Huge"
    elif runs>=50:
        return "Big"
    elif runs>0:
        return "Close"
    else:
        return "Chasing"
```

```python
df['Margin_Type'] = df['win_by_runs'].apply(margin_type)
df.head()
```

How many matches of each margin_type

```python
df['Margin_Type'].value_counts()
```

```python


data = {
    "Order_ID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Customer": ["A", "B", "A", "C", "B", "A", "C", "D"],
    "Category": ["Electronics", "Clothing", "Electronics", "Grocery", "Clothing", "Grocery", "Electronics", "Clothing"],
    "City": ["Bangalore", "Mumbai", "Bangalore", "Delhi", "Mumbai", "Bangalore", "Delhi", "Mumbai"],
    "Amount": [20000, 1500, 30000, 800, 2000, 500, 25000, 1800],
    "Quantity": [1, 3, 1, 5, 2, 4, 1, 2]
}

df2 = pd.DataFrame(data)

df2
```

Total sales for each category

```python
df2.groupby('Category')['Amount'].sum()
```

Average order value per city

```python
df2.groupby('City')['Amount'].mean()
```

Number of orders per customer

```python
df2.groupby('Customer')['Order_ID'].count()
```

Maximum order amount per category

```python
df2.groupby('Category')['Amount'].max()
```

Show multiple aggregations together

```python
df2.groupby('Category').agg(
    {'Amount' : ['sum', 'mean', 'min', 'max'],
     'Quantity' : ['sum', 'min', 'max']})
```
