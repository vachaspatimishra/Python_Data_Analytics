Break statement (break loop)

```python
for i in range(1,11):
    if i == 6:
        break
    else:
        print(i)
```

Continue statement (skip)

```python
for i in range(1,11):
    if i == 5:
        continue
    else:
        print(i, end=", ")
```

Pass statement (placeholder-do not do anything)

```python
for i in range(1,11):
    pass
```

While loop

```python
password = ""
while(password != "python"):
    password = input("Enter the password")
print("Logged in")
```

List Comprehension

Create a list of squares from 1 to 10

```python
for i in range(1,11):
    print(i**2, "   ", i**3)
```

```python
[i**2 for i in range(1,11)]
```

Create a list of even numbers from 1 to 20

```python
[i for i in range(1, 21) if i%2 == 0]
```

Create a list of odd numbers from 1 to 20

```python
[i for i in range(1,21) if i%2 ==1]
```

```python
[i for i in range(1,21) if i%2 !=0]
```

Convert all names in a list to uppercase

```python
a = ["Apple", "Mango", "Banana"]
[i.upper() for i in a]
```

Create a list of lengths of each word in ["cat", "elephant", "dog"]

```python
b = ["cat", "elephant", "dog"]
[len(i) for i in b]
```

From a list of numbers, create another list containing only numbers > 10

```python
e = [5,10,20,12]
```

```python
[i for i in e if i>10]
```
