Check if the number is divisible by 3 & 5, only 3, only 5

```python
n = int(input("Enter the number :"))
if(n % 3 == 0 and n % 5 == 0):
    print(f"{n} is divisible by both 3 and 5")
elif(n % 3 == 0):
    print(f"{n} is divisible by 3")
elif(n % 5 == 0):
    print(f"{n} is divisible by 5")
else:
    print(f"{n} is not divisible by 3 or 5")
```

Find sum of numbers from 1 to n.

```python
num = int(input("Enter the number :"))
sum = 0
for i in range (1,num+1):
    sum = sum + i
print(f"Sum of the numbers is {sum}")
```

Count number of vowels in a string.

```python
text = input("Enter the name :")
count = 0
for i in text:
    if (i in "aeiou"):
        count = count+1
print("Number of vowels :", count)
```

```python
text = input("Enter the name :")
count = 0
for i in text:
    if (i in "aeiouAEIOU"):
        count = count+1
print("Number of vowels :", count)
```
