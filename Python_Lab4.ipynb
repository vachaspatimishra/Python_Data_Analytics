# NumPy: Aggregations

**Exercise 1:** Create `arr = np.arange(1,10).reshape(3,3)` and print the overall `sum()`, `mean()`, and `std()`.

```python
arr = np.arange(1,10).reshape(3,3)
print(arr)
print(arr.sum())
print(arr.mean())
print(arr.std())
```

```python
arr = np.arange(1,10).reshape(3,3)
print(arr)
print(np.sum(arr))
print(np.mean(arr))
print(round(np.std(arr), 2))
```

**Exercise 2:** Using the same `arr`, print column-wise sums (axis=0) and row-wise means (axis=1).

```python
print(np.sum(arr, axis=0))        # axis 0 - Column-wise
print(np.mean(arr, axis=1))       # axis 1 - Row-wise
print(np.mean(arr, axis=1).astype(int))        # Converted to integer
```
