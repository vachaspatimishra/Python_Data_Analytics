```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Amit', 'Sara', 'John', 'Priya', 'Ravi'],
    'Math': [80, 90, 75, 85, 60],
    'Science': [70, 95, 80, 75, 65],
    'Passed': [True, True, True, True, False]
})
```

```python
df
```

How many rows and columns are there?

```python
df.shape
```

What are the column names?

```python
df.columns
```

What is the data type of each column?

```python
df.dtypes
```

How will you select only the Math marks column?

```python
df['Math']
```

How will you select Name and Science columns together?

```python
df[['Name', 'Science']]
```

How will you get the first row?

```python
df.iloc[0]
```

How will you get the last row?

```python
df.iloc[-1]
```

How will you get the 2nd row and 3rd column value?

```python
print(df.iloc[1, 2])
```

How will you get first 3 rows?

```python
print(df.iloc[0:3])
```

How will you get rows 2 to 4 and columns 1 to 2?

```python
print(df.iloc[1:4, 0:2])
```

How will you get all students who passed?

```python
df[df['Passed'] == True]
```

How will you get students with Math > 80 AND Science > 80?

```python
df[(df['Math']>80) & (df['Science']>80)]
```

How will you create a new column Total = Math + Science?

```python
df['Total'] = df['Math'] + df['Science']
df
```

How will you create Average marks?

```python
df['Average'] = df['Total']/2
df
```

Above 80 → Excellent, else Average. Create new column called category

```python
def remark(marks):
    if marks>80:
        return "Excellent"
    else:
        return "Average"
df["Category"] = df["Average"].apply(remark)
df
```

How will you label Passed column:

True → "Yes", False → "No"

```python
def grade(result):
    if result == True:
        return "Yes"
    else:
        return "No"
df['Label'] = df['Passed'].apply(grade)
df
```

How will you rename column Math → Maths?

```python
df.rename(columns = {'Math':'Maths'}, inplace = True)
df
```

How will you drop column Science?

```python
df.drop(columns = ['Science'], inplace = True)
df
```

What is the average Math score of class?

```python
print(df['Maths'].mean())
```

How many students passed?

```python
print(df['Passed'].sum())
```
