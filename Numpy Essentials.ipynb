You have marks of students:

marks = [45, 78, 90, 66, 88, 72, 59]
Tasks:  
Convert to NumPy array  
Find average marks  
Find highest & lowest marks  
Find students scoring above average  
Add 5 grace marks to all students

```python
marks = np.array([45, 78, 90, 66, 88, 72, 59])
print("Average is :", np.mean(marks))
print("Highest Marks :", np.max(marks))
print("Lowest Marks :", np.min(marks))
print("After adding 5 grace marks :", marks + 5)
print("Above average marks :", marks[marks>np.mean(marks)])       # Filtering
```

A grayscale image (3x3 pixels):

img = [[100, 150, 200],
       [50, 75, 125],
       [0, 25, 255]]
Tasks:  
Convert to NumPy array  
Increase brightness by 20  
Clip values to max 255  

```python
img = np.array([[100, 150, 200], [50, 75, 125], [0, 25, 255]])
print(img)
print("Increased brightness :", img + 20)
clipped = np.clip(img, 0, 255)
print("Clipped values to 255 :", clipped)
```

Sales for 5 days:

sales = [200, 300, 250, 400, 350]
Tasks:  
Find total sales  
Find day with highest sales  
Find days where sales < 300  

```python
sales = np.array([200, 300, 250, 400, 350])
print("Total sales :", np.sum(sales))
print("Highest sales :", np.max(sales))
print("Sales < 300 :", sales<300)
```
