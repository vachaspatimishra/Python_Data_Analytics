Select players who scored more than 500 runs.

```python
df.loc[df["Runs"]>500, ["Player Name", "Runs"]]
```

Select "Player Name" and "SR" where SR > 150 using loc.

```python
df.loc[df["SR"]>150, ["Player Name", "SR"]]
```

Select players from team "RCB".

```python
df.loc[df["Team"] == "RCB", ["Player Name", "Team"]]
```

Select players where Runs > 600 and display only "Player Name" and "Team".

```python
df.loc[df["Runs"] > 600, ["Player Name", "Team"]]
```

Select players where 50s >= 5.

```python
df.loc[df["50s"] > 5, ["Player Name"]]
```

Select players with at least one century (100s > 0).

```python
df.loc[df["100s"] > 0, ["Player Name"]]
```

Select players where Runs > 500 AND SR > 140.

```python
df.loc[(df["Runs"] > 500) & (df["SR"] > 140), ["Player Name"]]
```

Select players where Team = "MI" AND Runs > 400.

```python
df.loc[(df["Team"] == "MI") & (df["Runs"] > 400), ["Player Name"]]
```

Sort players by Runs in descending order.

```python
df.sort_values(by = 'Runs', ascending=False)[['Player Name', 'Runs']]
```

```python
df[['Player Name', 'Runs']].sort_values(by='Runs', ascending=False)       # [[]] used for multiple columns
```

**Why use `&` instead of 'and'?**

1. & operator works for each row (Element-wise comparison).
2. python works on c++ in backend and c++ supports & operator.
3. 'and' for single boolean values

and - "Is this ONE thing true?"
& - "Check ALL rows one by one"
