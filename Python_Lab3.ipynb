# Exceptions: try/except, else, finally; Common Errors; raise

**Exercise 1:** Wrap a division `a/b` in try/except to catch `ZeroDivisionError` and print a helpful message.

```python
a = 5
b = 0
try:
    result = a/b
except:
    print("Division by zero error")
else:
    print(result)
finally:
    print("Division Operation")
```

```python
a = 5
b = 0
try:
    a/b
except ZeroDivisionError as e:
    print(e)
```

**Exercise 2:** Convert the string `'42'` to int inside try/except/else/finally; print success in else, 'done' in finally.

```python
try:
    print(int("42"))
except:
    print("Error occured")
else:
    print("Success")
finally:
    print("Done...")
```

**Exercise 6:** Define a custom exception class `TooSmallError`. Write a function that raises it if input < 10; test with 7 and handle it.

```python
class TooSmallError(Exception):
    pass
try:
    a = int(input("Enter the number :"))
    if a<10:
        raise TooSmallError("The number you entered is too small")
except TooSmallError as e:
    print(e)
```

**Exercise 7:** Read an integer from a string; if conversion succeeds (else branch), print its square; otherwise print 'bad input'. Use `s='16'`.

```python
class ConversionError(Exception):
    pass
s = "16"
try:
    int(s)
    raise ConversionError("Bad input")
except ConversionError as e:
    print(e)
else:
    print("Square of the number is :", s**2)
```

**Exercise 5:** From a list `['red','green','blue','yellow']`, randomly shuffle the list and print it. Then sample 2 unique colors.

```python
colors = ['red','green','blue','yellow']
random.shuffle(colors)
print(random.choices(colors, k=3))
print(random.sample(colors, 2))
```

**Exercise 9:** Parse `'2025-09-11 08:30:00'` to a `datetime`, add 90 minutes, and print the result in `YYYY-MM-DD HH:MM` format.

```python
from datetime import timedelta, datetime
dt = datetime.strptime('2025-09-11 08:30:00', "%Y-%m-%d %H:%M:%S")
print(dt)
future = dt+timedelta(minutes=90)
print(future)
```

**Exercise 10:** Create a small schedule: start at today's date and print the next 5 day names (Mon..Sun) using `datetime`.

```python
from datetime import datetime, timedelta

# Get today's date
today = datetime.today()

# Loop for next 5 days
for i in range(1,6):
    day = today + timedelta(days=i)
    print(day.strftime("%A"))
```

**After months**

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta
today = datetime.today()
for i in range(1,6):
    nextmonth = today + relativedelta(months=i)
    print(nextmonth.strftime("%B"))
```
