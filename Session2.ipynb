```python
name = input("Please enter your name :")
year = input("Enter the year")
print(f"Hello {name}! You are learning python in {year}")
```

```python
test_scores  = [45, 78, 90, 66]
# Find total marks, highest, lowest, and average.

print(f"Total marks is {sum(test_scores)}")
print(f"Highest marks is {max(test_scores)}")
print(f"Lowest marks is {min(test_scores)}")
print(f"Average of marks is {sum(test_scores)/len(test_scores)}")
```

```python
score = int(input("Enter your score : "))
if(score > 90):
    print("Excellent")
elif(score > 60):
    print("Good")
else:
    print("Try Again!")
```

```python
temp = int(input("Enter the temperature"))
if(temp > 30):
    print("Hot")
elif(temp > 20):
    print("Nice Weather")
else:
    print("Cold")
```

```python
# If your total bill crosses ₹500, you get a 10% discount.
bill = int(input("Enter the bill amount : "))
if(bill > 500):
    print(f"Total amount after 10% discount is {bill-bill*0.1}")
else:
    print(bill)
```

```python
# >90 --> Excellent, >=75 --> Good, >=50 --> Improvement, else F
marks = int(input("Enter your marks : "))
if(marks > 90):
    print("Excellent")
elif(marks >= 75):
    print("Good")
elif(marks >= 50):
    print("Improvement Needed")
else:
    print("Failed")
```
