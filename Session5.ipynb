```python
def math():
    print("***Calculator***")
    num1 = int(input("Enter the first number :"))
    num2 = int(input("Enter the second number :"))
    cal = int(input("Enter your choice >>> \n Addition : Type 1 \n Subtraction : Type 2 \n Multiplication : Type 3 \n Division : Type 4 \n"))
    if cal==1:
        print("Addition of the numbers is :", num1+num2)
    elif cal==2:
        print("Subtraction of the numbers is :", num1-num2)
    elif cal==3:
        print("Multiplication of the numbers is :", num1*num2)
    elif cal==4:
        print("Division of the numbers is :", num1/num2)
    else:
        print("Invalid Input")
```

```python
def maths():
    while True:
        print("\n ***Calculator***")
        num1 = int(input("Enter the first number :"))
        num2 = int(input("Enter the second number :"))
        cal = int(input("\n Enter your choice >>> \n"
                        "Addition : Type 1 \n"
                        "Subtraction : Type 2 \n"
                        "Multiplication : Type 3 \n"
                        "Division : Type 4 \n"))
        if cal==1:
            print("Addition of the numbers is :", num1+num2)
        elif cal==2:
            print("Subtraction of the numbers is :", num1-num2)
        elif cal==3:
            print("Multiplication of the numbers is :", num1*num2)
        elif cal==4:
            if num2!=0:
                print("Division of the numbers is :", num1/num2)
            else:
                print("Cannot divide by zero")
        else:
            print("Invalid Input")
        choice = input("\n Do you want to continue? (y/n) :").lower()
        if choice != "y":
            print("---Calculator closed---")
            break
```
