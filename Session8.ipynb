Divide the two numbers and print the result only if there is no error. At the end print "Division operation"

```python
a = int(input("Enter the first number :"))
b = int(input("Enter the second number :"))
try:
    result = a/b
except:
    print("Division by zero error")
else:
    print(result)
finally:
    print("Division Operation")
```

Read a text file, print the contents of the file only if there is no error. At the end print "File Handling Done"

```python
try:
    with open(r"abc.text", "r") as f:
        content = f.read()
except:
    print("File Not Found")
else:
    print(content)
finally:
    print("...File Handling Done")
```

Take number as input from the user, print square if input is valid, else print error

```python
try:
    a = int(input("Enter the number :"))
except:
    print("Invalid input")
else:
    print(f"Square of {a} is : {a**2}")
finally:
    print("Squaring is done...")
```

Random Module

```python
import random
print(random.randint(1,6))
```

```python
print(random.randrange(1,11,2))      #Odd numbers
```

```python
print(random.randrange(0,11,2))      #Even numbers
```

```python
fruits = ["Apple", "Mango", "Banana", "Strawberry"]
print(random.choice(fruits))
```

```python
fruits = ["Apple", "Mango", "Banana", "Strawberry"]
print(random.choices(fruits, k=3))
```
