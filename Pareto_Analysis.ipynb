# Pareto Analysis - 80/20 Rule

**80% Revenue from 20% Customers**

```python
import pandas as pd

data = {
        'Customers': ['A', 'B', 'C' , 'D' , 'E'],
        'Monetary': [800, 1000, 200, 500, 100]
}
```

## Creating DataFrame

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

## Sorting Monetary in descending order

```python
df = df.sort_values(by = 'Monetary', ascending=False)
df
```

## Cumulative Revenue - adds all the revenue from start to a specific point

```python
df['Cumulative_Revenue'] = df['Monetary'].cumsum()
df
```

## Revenue Percentage

```python
df['Revenue_Percent'] = df['Cumulative_Revenue']/df['Monetary'].sum() * 100
df
```

## Resetting the index for proper sequence (important to calculate customer percentage)

```python
df.reset_index(drop=True)
```

## Customer percentage

```python
df['Customer_Percent'] = (df.index+1)/len(df) * 100
df
```

**This dataset does NOT satisfy the pareto analysis as 20% customers contribute only about 70% revenue.**
